```python
from netCDF4 import Dataset
import pandas as pd
import numpy as np
import os
import datetime
import tqdm
from ll_xy import lonlat_to_xy
import pickle
import matplotlib.pyplot as plt

sic_dir = '/home/robbie/uit_mnt/Data/romal7177/ResearchData/IFT/EarthObservation/SatelliteAltimetry/OSISAF Sea Ice Concentration'

ice_cover_dir = '/home/robbie/uit_mnt/home/romal7177/arctex/ice_cover/'
```

```python
frames = {}

for radius in [100,200]:

    list_of_dfs = []

    for year in tqdm.trange(2003,2024):
        
        df = pickle.load(open(f'{ice_cover_dir}{radius}/cover_{year}.p','rb'))

        df['doy'] = [x.timetuple().tm_yday for x in df['dt']]

        list_of_dfs.append(df)

    df = pd.concat(list_of_dfs)

    max_cover = np.nanmax(df['cover'])

    df['norm'] = df['cover']/max_cover

    df.sort_values('dt',inplace=True)
    
    frames[radius] = df
```

```python
increment=7

for radius in [100,200]:
    
    df = frames[radius]

    for start_year in np.arange(2003,2024,increment):

    #     print(start_year,start_year+5)

        df_period = df[(df['year']>=start_year)&(df['year']<start_year+increment)]

        df_period.drop(columns=['dt','month','year','day'],inplace=True)

        df_period = df_period.groupby('doy').mean()

        df_period = df_period[df_period.index<366]

        plt.plot(df_period.index,df_period['norm'],label=f'{start_year} - {start_year + increment}')


    plt.ylim(0,1)
    plt.ylabel('Normalised Ice Cover',fontsize='x-large')
    plt.xlabel('Day of Year',fontsize='x-large')
    plt.title(f'Austfonna Surrounding Ice Cover ({thresh} km)',fontsize='xx-large')
    plt.xlim(0,365)
    plt.legend(title='Period')
    plt.show()
```

```python
year=2003
f = 'ice_conc_nh_ease2-250_cdr-v2p0_200303101200.nc'

d = Dataset(f'{sic_dir}/{year}/{f}')

xgrid,ygrid=lonlat_to_xy(lons,lats,hemisphere='n')
aust_lon, aust_lat = 24.6,79.8
aust_x,aust_y = lonlat_to_xy(np.array(aust_lon),np.array(aust_lat),hemisphere='n')

distances_to_austfonna = np.sqrt((xgrid-aust_x)**2+(ygrid-aust_y)**2)

lons = np.array(d['lon'])
lats = np.array(d['lat'])

dist_thresh=100_000

list_of_dicts = []

for year in np.arange(2004,2024):
    print(year)
    
    files = sorted(os.listdir(f'{sic_dir}/{year}'))
    
    
    for f in tqdm.tqdm(files):
        datestr = f.split('_')[-1][-15:-7]
        year = int(datestr[:4])
        month = int(datestr[4:6])
        day = int(datestr[6:8])
        dt = datetime.date(year,month,day)
        
        with Dataset(f'{sic_dir}/{year}/{f}') as d:

            sic = np.array(d['ice_conc'])[0]/100
            sic[sic<0]=np.nan
            sic_nearby = sic.copy()
            sic_nearby[distances_to_austfonna>dist_thresh]=np.nan

            total_sic = np.nansum(sic_nearby)

            dic = {'cover':total_sic,'year':year,'month':month,'day':day,'dt':dt}

            list_of_dicts.append(dic)
        
    df = pd.DataFrame(list_of_dicts)
    
    pickle.dump(df,open(f'cover_{year}.p','wb'))
```
